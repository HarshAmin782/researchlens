```python
import pickle
import json
import re
import time
import numpy as np
from abc import ABC, abstractmethod
from typing import List, Dict, Tuple
import chromadb
from sentence_transformers import SentenceTransformer, CrossEncoder
from rank_bm25 import BM25Okapi
```

```python
with open('/kaggle/working/bm25_index.pkl', 'rb') as f:
    bm25_index = pickle.load(f)

with open('/kaggle/working/bm25_metadata.pkl', 'rb') as f:
    bm25_metadata = pickle.load(f)

chroma_client = chromadb.PersistentClient(path='/kaggle/working/chroma_db')
fixed_collection = chroma_client.get_collection("fixed_chunks")
semantic_collection = chroma_client.get_collection("semantic_chunks")

embed_model = SentenceTransformer('all-MiniLM-L6-v2')
reranker = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')

print(f"BM25: {len(bm25_metadata)} docs")
print(f"ChromaDB fixed: {fixed_collection.count()}")
print(f"ChromaDB semantic: {semantic_collection.count()}")
print(f"Models loaded")
```

```python
class BaseRetriever(ABC):
    @abstractmethod
    def retrieve(self, query: str, top_k: int = 5) -> List[Dict]:
        pass

    @property
    @abstractmethod
    def strategy_name(self) -> str:
        pass

    def _make_result(self, text, score, paper_id, title, chunk_id):
        return {
            'text': text,
            'score': float(score),
            'paper_id': paper_id,
            'title': title,
            'chunk_id': chunk_id
        }

class NaiveRetriever(BaseRetriever):
    @property
    def strategy_name(self):
        return "naive_rag"

    def retrieve(self, query: str, top_k: int = 5) -> List[Dict]:
        query_embedding = embed_model.encode(query).tolist()
        results = fixed_collection.query(
            query_embeddings=[query_embedding],
            n_results=top_k,
            include=['documents', 'distances', 'metadatas']
        )
        return [self._make_result(
            doc, 1 - dist, meta['paper_id'], meta['title'], meta.get('chunk_id', '')
        ) for doc, dist, meta in zip(
            results['documents'][0], results['distances'][0], results['metadatas'][0]
        )]

class SemanticRetriever(BaseRetriever):
    @property
    def strategy_name(self):
        return "semantic_rag"

    def retrieve(self, query: str, top_k: int = 5) -> List[Dict]:
        query_embedding = embed_model.encode(query).tolist()
        results = semantic_collection.query(
            query_embeddings=[query_embedding],
            n_results=top_k,
            include=['documents', 'distances', 'metadatas']
        )
        return [self._make_result(
            doc, 1 - dist, meta['paper_id'], meta['title'], meta.get('chunk_id', '')
        ) for doc, dist, meta in zip(
            results['documents'][0], results['distances'][0], results['metadatas'][0]
        )]

class HybridRetriever(BaseRetriever):
    @property
    def strategy_name(self):
        return "hybrid_rag"

    def retrieve(self, query: str, top_k: int = 5) -> List[Dict]:
        fetch_k = top_k * 4
        query_embedding = embed_model.encode(query).tolist()
        dense_results = fixed_collection.query(
            query_embeddings=[query_embedding],
            n_results=fetch_k,
            include=['documents', 'distances', 'metadatas']
        )
        dense_hits = {}
        for rank, (doc, dist, meta) in enumerate(zip(
            dense_results['documents'][0], dense_results['distances'][0], dense_results['metadatas'][0]
        )):
            chunk_id = meta.get('chunk_id', f"dense_{rank}")
            dense_hits[chunk_id] = {'rank': rank, 'text': doc, 'score': 1 - dist,
                                     'paper_id': meta['paper_id'], 'title': meta['title']}

        bm25_scores = bm25_index.get_scores(query.lower().split())
        top_bm25_idx = np.argsort(bm25_scores)[::-1][:fetch_k]
        bm25_hits = {}
        for rank, idx in enumerate(top_bm25_idx):
            chunk_id = bm25_metadata[idx]['chunk_id']
            bm25_hits[chunk_id] = {'rank': rank, 'text': bm25_metadata[idx]['text'],
                                    'score': bm25_scores[idx], 'paper_id': bm25_metadata[idx]['paper_id'],
                                    'title': bm25_metadata[idx]['title']}

        k = 60
        all_chunk_ids = set(dense_hits.keys()) | set(bm25_hits.keys())
        rrf_scores = {}
        for chunk_id in all_chunk_ids:
            rrf = 0.0
            if chunk_id in dense_hits:
                rrf += 1 / (dense_hits[chunk_id]['rank'] + k)
            if chunk_id in bm25_hits:
                rrf += 1 / (bm25_hits[chunk_id]['rank'] + k)
            rrf_scores[chunk_id] = rrf

        top_ids = sorted(rrf_scores, key=rrf_scores.get, reverse=True)[:top_k]
        output = []
        for chunk_id in top_ids:
            source = dense_hits.get(chunk_id) or bm25_hits.get(chunk_id)
            output.append(self._make_result(
                source['text'], rrf_scores[chunk_id],
                source['paper_id'], source['title'], chunk_id
            ))
        return output

class HybridRerankerRetriever(BaseRetriever):
    @property
    def strategy_name(self):
        return "hybrid_rerank"

    def retrieve(self, query: str, top_k: int = 5) -> List[Dict]:
        candidates = HybridRetriever().retrieve(query, top_k=20)
        pairs = [[query, c['text']] for c in candidates]
        rerank_scores = reranker.predict(pairs)
        for candidate, score in zip(candidates, rerank_scores):
            candidate['score'] = float(score)
        return sorted(candidates, key=lambda x: x['score'], reverse=True)[:top_k]

print("All retriever classes defined")
```

```python
QUERY_SIGNALS = {
    'consensus': [
        'do papers agree', 'do researchers agree', 'is there consensus',
        'what do papers say', 'controversy', 'debate', 'conflicting',
        'disagree', 'contradiction', 'contradictory', 'dispute'
    ],
    'comparative': [
        'compare', 'difference between', 'differ', 'versus', ' vs ',
        'better than', 'worse than', 'advantage', 'disadvantage',
        'trade-off', 'tradeoff', 'which is', 'contrast'
    ],
    'procedural': [
        'how to', 'how do i', 'steps to', 'implement', 'build',
        'create', 'set up', 'train', 'fine-tune', 'walkthrough', 'tutorial'
    ],
    'factual': [
        'what is', 'what are', 'define', 'definition', 'explain',
        'describe', 'who is', 'when was', 'where is', 'which paper'
    ]
}

QUERY_TO_STRATEGY = {
    'factual': 'naive_rag',
    'comparative': 'hybrid_rerank',
    'consensus': 'hybrid_rag',
    'procedural': 'semantic_rag'
}

def keyword_classify(query: str) -> Tuple[str, float]:
    query_lower = query.lower()
    for qtype, signals in QUERY_SIGNALS.items():
        for signal in signals:
            if signal in query_lower:
                return qtype, 0.95
    return 'factual', 0.60

print("Query taxonomy defined")
print(f"Types: {list(QUERY_SIGNALS.keys())}")
print(f"Strategy mapping: {QUERY_TO_STRATEGY}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "mistralai/Mistral-7B-Instruct-v0.2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
llm = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
print("Mistral loaded")
```

```python
def llm_classify(query: str) -> Tuple[str, float]:
    prompt = f"""[INST] Classify this research question into exactly one category.

Categories:
- factual: asking for a definition, explanation, or specific fact
- comparative: asking to compare two or more things
- consensus: asking whether papers agree or disagree on something
- procedural: asking how to implement or do something step by step

Question: {query}

Respond with exactly one word — the category name. [/INST]"""

    inputs = tokenizer(prompt, return_tensors="pt").to(llm.device)

    with torch.no_grad():
        outputs = llm.generate(
            **inputs,
            max_new_tokens=10,
            temperature=0.1,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )

    response = tokenizer.decode(
        outputs[0][inputs['input_ids'].shape[1]:],
        skip_special_tokens=True
    ).strip().lower()

    for qtype in QUERY_SIGNALS.keys():
        if qtype in response:
            return qtype, 0.85

    return 'factual', 0.50

print("LLM classifier defined")
```

```python
def route_query(query: str, llm_fallback_threshold: float = 0.7) -> Dict:
    qtype, confidence = keyword_classify(query)

    classifier_used = 'keyword'
    if confidence < llm_fallback_threshold:
        qtype, confidence = llm_classify(query)
        classifier_used = 'llm'

    strategy_name = QUERY_TO_STRATEGY[qtype]
    strategy_map = {
        'naive_rag': NaiveRetriever(),
        'semantic_rag': SemanticRetriever(),
        'hybrid_rag': HybridRetriever(),
        'hybrid_rerank': HybridRerankerRetriever()
    }

    return {
        'query': query,
        'query_type': qtype,
        'strategy': strategy_name,
        'confidence': confidence,
        'classifier_used': classifier_used,
        'retriever': strategy_map[strategy_name]
    }

print("Router defined")
```

```python
test_queries = [
    "What is LoRA?",
    "How does attention mechanism work in transformers?",
    "Compare LoRA and full fine-tuning approaches",
    "Do papers agree on the optimal learning rate for fine-tuning?",
    "How do I implement a RAG pipeline from scratch?",
    "What are the limitations of retrieval augmented generation?",
    "Is there consensus on whether larger models are always better?",
    "Explain the difference between encoder and decoder transformers"
]

print(f"{'Query':<55} {'Type':<12} {'Strategy':<16} {'Conf':<6} {'Via'}")
print("-" * 105)

for query in test_queries:
    result = route_query(query)
    print(f"{query[:54]:<55} {result['query_type']:<12} {result['strategy']:<16} {result['confidence']:<6.2f} {result['classifier_used']}")
```

```python
labeled_queries = [
    ("What is transformer architecture?", "factual"),
    ("Define attention mechanism", "factual"),
    ("What is BERT?", "factual"),
    ("Explain positional encoding", "factual"),
    ("What are embeddings?", "factual"),
    ("What is gradient descent?", "factual"),
    ("What does RAG stand for?", "factual"),
    ("What is a language model?", "factual"),
    ("What is perplexity in NLP?", "factual"),
    ("What is tokenization?", "factual"),
    ("Compare BERT and GPT architectures", "comparative"),
    ("What is the difference between LoRA and QLoRA?", "comparative"),
    ("LoRA vs full fine-tuning tradeoffs", "comparative"),
    ("Which is better, BM25 or dense retrieval?", "comparative"),
    ("Contrast encoder-only and decoder-only models", "comparative"),
    ("Compare RAG versus fine-tuning for knowledge injection", "comparative"),
    ("What are the advantages of sparse vs dense retrieval?", "comparative"),
    ("Difference between self-attention and cross-attention", "comparative"),
    ("How do retrieval methods differ in accuracy?", "comparative"),
    ("Compare chunking strategies for RAG", "comparative"),
    ("Do papers agree on optimal chunk size for RAG?", "consensus"),
    ("Is there consensus on dropout rates in transformers?", "consensus"),
    ("Do researchers agree that scaling always helps?", "consensus"),
    ("Is there debate about positional encoding methods?", "consensus"),
    ("Do papers contradict each other on learning rate schedules?", "consensus"),
    ("What is the controversy around attention head pruning?", "consensus"),
    ("Is there conflicting evidence on RAG vs fine-tuning?", "consensus"),
    ("Do papers disagree about optimal context window size?", "consensus"),
    ("Is there consensus on the best evaluation metric for RAG?", "consensus"),
    ("Are there contradictory findings on temperature in generation?", "consensus"),
    ("How to implement LoRA from scratch?", "procedural"),
    ("Steps to build a RAG pipeline", "procedural"),
    ("How do I fine-tune a language model?", "procedural"),
    ("How to set up ChromaDB for semantic search?", "procedural"),
    ("How to train a cross-encoder reranker?", "procedural"),
    ("Steps to evaluate a RAG system with RAGAS", "procedural"),
    ("How to implement BM25 keyword search?", "procedural"),
    ("How do I chunk documents for retrieval?", "procedural"),
    ("How to implement reciprocal rank fusion?", "procedural"),
    ("Steps to deploy a Streamlit app to the cloud", "procedural"),
]

correct = 0
wrong_cases = []

for query, expected in labeled_queries:
    result = route_query(query)
    predicted = result['query_type']
    if predicted == expected:
        correct += 1
    else:
        wrong_cases.append({
            'query': query,
            'expected': expected,
            'predicted': predicted,
            'confidence': result['confidence'],
            'via': result['classifier_used']
        })

accuracy = correct / len(labeled_queries)
print(f"Routing accuracy: {correct}/{len(labeled_queries)} = {accuracy:.1%}")
print(f"\nMisclassified ({len(wrong_cases)}):")
for w in wrong_cases:
    print(f"  [{w['expected']} → {w['predicted']}] ({w['via']}, {w['confidence']:.2f}) {w['query'][:70]}")
```

```python
def route_and_retrieve(query: str, top_k: int = 5) -> Dict:
    routing = route_query(query)
    retriever = routing['retriever']

    start = time.time()
    chunks = retriever.retrieve(query, top_k=top_k)
    latency = time.time() - start

    return {
        'query': query,
        'query_type': routing['query_type'],
        'strategy': routing['strategy'],
        'confidence': routing['confidence'],
        'chunks': chunks,
        'latency_ms': round(latency * 1000)
    }

demo_queries = [
    "What is LoRA?",
    "Compare dense and sparse retrieval methods",
    "How do I implement attention from scratch?",
    "Do papers agree on whether RAG beats fine-tuning?"
]

for query in demo_queries:
    result = route_and_retrieve(query)
    print(f"Query:    {result['query']}")
    print(f"Type:     {result['query_type']} → {result['strategy']} ({result['latency_ms']}ms)")
    print(f"Top chunk: {result['chunks'][0]['text'][:200]}")
    print()
```

```python
routing_eval = {
    'accuracy': accuracy,
    'total': len(labeled_queries),
    'correct': correct,
    'wrong_cases': wrong_cases
}

with open('/kaggle/working/routing_eval.json', 'w') as f:
    json.dump(routing_eval, f, indent=2)

print("Saved routing_eval.json")
print(f"\n=== Notebook 04 Complete ===")
print(f"Routing accuracy: {accuracy:.1%}")
print(f"Router: keyword classifier + Mistral LLM fallback")
```
